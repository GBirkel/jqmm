# Yeast on glucose

Data for two conditions is stored here: growth on labeled glucose and growth on labeled ethanol. **SBML files** can be obtained through the getSBMLfile method (inspect the function to see the details):

```python
GlcFile = YD.getSBMLfile('Glc')  
```

Similarly, one can get the two-scale model through the getTSmodel function:

```python
GlcTSmodel = YD.getTSmodel('Glc')
```

or the actual results with getTSresults:

```python
%%time
TSresult,TSmodel = YD.getTSresults('Glc')
```

Once you have the results, it is easy to plot the labeling data fits:

```python
TSresult.plotExpvsCompLabelFragment(titleFig='wyr1Glc')
```

Or draw the fluxes on the svg map: 

```python
TSresult.drawFluxes('YeastGlc.svg',svgInFileName='SciMM904TKs.svg')      
SVG('YeastGlc.svg')
```

Better yet, you can write down the result in an SBML file:

```python
TSresult.reactionNetwork.write('SMM904Glc.sbml')
```

and upload it to mvt.jbei.org

We can also get the External Labeling Variability Analysis (ELVA):

```python
TSresult2  = TSmodel.ELVA(TSresult, erase=False)  
TSresult2.plotExpvsCompLabelxvsy(titleFig='Glc',outputFileName="ELVAComparisonGlc.txt")
```
