# Calculations

Load the data from the repository in the library:

```python
filename = '/users/hgmartin/libraries/quantmodel/data/tests/PCAP/PCAP_Limonene5.csv'
PCAP = PCAPlib.dataSet(filename,product='Limonene',order=['AtoB','HMGS','HMGR','MK','PMK','PMD','Idi','GPPS','LS'])
```

Plotting data as a heatmap:

```python
PCAP.plotData()
```

```python
print PCAP.Z
```

Do PCAP analysis not using BL-Mm for calculating principal components, but plotting its projection.

These are figures 3a, 3b, 3c and 4c in Alonso-Gutierrez *et al*:

```python
PCAP.plotPCAP(eliminate=['BL-Mm'],showEliminate=True)
```
